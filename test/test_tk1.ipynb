```python
import threading
import tkinter as tk
class Threader(threading.Thread):
    def __init__(self, *args, **kwargs):
            
            threading.Thread.__init__(self, *args, **kwargs)
            self.daemon = True
            self.start()
    def run(self):
    #要執行的task
    #main thread
        if __name__ == '__main__':
            
            window=tk.Tk()
            window.title('查詢關鍵字新聞')
            window.geometry('450x300')
            window.resizable(0,0)
            tk.Label(window, text='請輸入欲查詢的關鍵字：').place(x=40, y= 45)
            tk.Label(window, text='請輸入日期區間之起始值：').place(x=40, y= 95)
        #GUI介面設計
            tk.Button(window, text='查詢', command= lambda: Threader(name='Thread-name')).place(x=150 , y=220, height=30, width=150) #新增thread
            threading.active_count() #查看thread總數量
            threading.enumerate() #列出所有thread name
            threading.current_thread() #查看目前進入哪個thread         
            
            window.mainloop()
```

```python
import random

def generate_lottery_numbers():
    # 生成 6 個隨機號碼，範圍在 1 到 49 之間
    numbers = random.sample(range(1, 50), 6)
    # 生成一個特別號碼，範圍在 1 到 49 之間
    special_number = random.choice(range(1, 50))
    return numbers, special_number

# 生成大樂透號碼
lottery_numbers, special_number = generate_lottery_numbers()
print("大樂透號碼:", lottery_numbers)
print("特別號碼:", special_number)
```
